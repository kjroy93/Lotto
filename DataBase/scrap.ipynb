```python
import requests
from selectolax.parser import HTMLParser
from time import gmtime
from dataclasses import dataclass

@dataclass(slots=True)
class WebData:
    age: int
    html: int

@dataclass
class LineaTabla:
    number: int
    date: None
    numero1: int
    numero2: int
    numero3: int
    numero4: int
    numero5: int
    estrella1: int
    estrella2: int


class EuroScraping:
    headers = {"User-gent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"}
    url = 'https://www.euromillones.com.es/historico/resultados-euromillones-{}.html'

    def __init__(self) -> None:
        self.content = []
        self.__get_data()

    def __get_data(self):
        with requests.Session() as s:
            for age in range(2004,gmtime().tm_year+1):
                resp = s.get(url=EuroScraping.url.format(age),headers=EuroScraping.headers).text
                self.content.append(WebData(age,resp))
    
    def parse_data(self):
        self.content

    def __str__(self) -> str:
        return self.content
    
    def __repr__(self) -> str:
        return 'Srapping done'
```
